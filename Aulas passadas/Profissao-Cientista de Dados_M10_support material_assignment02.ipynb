````python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
````

````python
df = pd.read_csv('previsao_de_renda.csv')
````

````python
df.columns
````

````python
df.head()
````

````python
plt.figure(figsize=(10, 6))
sns.pointplot(x='posse_de_imovel', y='renda', data=df, ci=None)
plt.xlabel('Posse de Imóvel')
plt.ylabel('Renda')
plt.title('Relação entre Posse de Imóvel e Renda')
plt.show()
````

````python
plt.figure(figsize=(10, 6))
sns.pointplot(x='posse_de_veiculo', y='renda', data=df, ci=None)
plt.xlabel('posse_de_veiculo')
plt.ylabel('Renda')
plt.title('Relação entre posse_de_veiculo e Renda')
plt.show()
````
